```python
from vantage6.algorithm.tools.mock_client import MockAlgorithmClient

client = MockAlgorithmClient(
    datasets=[
        [
            {
                "database": {
                    "dbA": converted_df,
                    "dbB": converted_df
                },
                "db_type": "omop"
            },
        ],
        [
            {
                "database": {
                    "dbA": converted_df,
                    "dbB": converted_df
                },
                "db_type": "omop"
            },
        ]
    ],
    module="v6-analytics",
)
```

```python
import psycopg2
import pandas as pd
```

```python
## Database related
def establish_connection():
    dbname = "omopdb"
    user = "ohdsi"
    password = "ohdsi"
    host = "localhost"
    port = "5432"

    conn = psycopg2.connect(
        dbname=dbname,
        user=user,
        password=password,
        host=host,
        port=port
    )

    cur = conn.cursor()
    return conn, cur
```

```python
with open('./query.sql', 'r') as file: query = file.read()
conn, cur = establish_connection()
cur.execute(query)
data = cur.fetchall()
column_names = [desc[0] for desc in cur.description]
df = pd.DataFrame(data, columns=column_names)
print(df)
```
